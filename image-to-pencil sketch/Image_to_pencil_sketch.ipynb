```python
# Importing Libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the image
image_path = "Khizer.jpg"
image = cv2.imread(image_path)
```

```python
# Original Image
plt.figure(figsize=(9, 9))
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('MY Original Image')
plt.axis('off')
plt.show()
```

```python
# Converting to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(9, 9))
plt.imshow(gray_image, cmap='gray')
plt.title('converted grayscale image')
plt.axis('off')
plt.show()
```

```python
# Inversion Elegance
inverted_image = cv2.bitwise_not(gray_image)
plt.figure(figsize=(7,7))
plt.imshow(inverted_image, cmap='gray')
plt.title('Inverted image')
plt.axis('off')
plt.show()
```

```python
# Gaussian Textures
blurred_image = cv2.GaussianBlur(inverted_image, (21, 21),0)
plt.figure(figsize=(8, 8))
plt.imshow(blurred_image, cmap='gray')
plt.title('Blurr Image')
plt.axis('off')
plt.show()
```

```python
inverted_blurred= cv2.bitwise_not(blurred_image)

# Display the pencil sketch image
pencil_sketch=cv2.divide(gray_image,inverted_blurred,scale=256.0)
plt.figure(figsize=(7, 7))
plt.imshow(pencil_sketch, cmap='gray')
plt.title('pencil sketch')
plt.axis('off')
plt.show()
```

```python
plt.figure(figsize=(20, 5))

# List of images
images = [image, gray_image, inverted_image, blurred_image, pencil_sketch]
titles = ["MY Original Image", "converted GrayScale Image", "Inverted Image", "Blurr Image", "output Pencil Sketch Image"]

# Loop to display images
for i in range(5):
    plt.subplot(1, 5, i + 1)
    plt.imshow(images[i],cmap='gray' if i != 0 else None )  # Display grayscale for non-original images
    plt.axis('off')
    plt.title(titles[i])

plt.tight_layout()
plt.show()
```
